# Create regulatory domains

This is an example of code allowing the use of the create_regdom function to create gene regulatory domains from a file containing the size of the chromosomes and a file containing the tss. Both files must be in .bed format documentation here: http://genome.ucsc.edu/FAQ/FAQformat#format1 

## For hg38

```python
%load_ext autoreload
%autoreload 2
```

```python
import greatpy as great
```

```python
regdom_hg38 = great.tl.REGDOM.create_regdom(
    tss_file="../data/human/hg38/tss_from_great.bed",
    chr_sizes_file="../data/human/hg38/chr_size.bed",
    association_rule="Basalplusextention",
    out_path=None
    )
regdom_hg38
```

## For hg19

```python
regdom_hg19 = great.tl.REGDOM.create_regdom(
    tss_file="../data/human/hg19/tss.bed",
    chr_sizes_file="../data/human/hg19/chr_size.bed",
    association_rule="Basalplusextention",
    out_path=None
    )
regdom_hg19
```
